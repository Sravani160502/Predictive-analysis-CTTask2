```python
import pandas as pd

# load dataset 

df = pd.read_csv('employees_dataset.csv')
```

```python
# Fill missing values in 'education' with mode


df['education'].fillna(df['education'].mode()[0], inplace=True) 
```

```python
# Fill missing values in 'previous_year_rating' with the median

df['previous_year_rating'].fillna(df['previous_year_rating'].median(), inplace=True)
```

```python
# save the data to new csv 

df.to_csv('New_employees_dataset.csv' , index=False)

print("missing value filled succesfully")
```

```python
# Define features and target
X = df.drop(['employee_id', 'KPIs_met_more_than_80'], axis=1)
y = df['KPIs_met_more_than_80']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=130,
                                   test_size=0.20,
                                   shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
le = LabelEncoder()

# Apply label encoding on each object (categorical) column
for col in X_train.columns:
    if X_train[col].dtype == 'object':
        X_train[col] = le.fit_transform(X_train[col])
        X_test[col] = le.transform(X_test[col])  # Important: transform, don't fit again
```

```python
from sklearn.model_selection import GridSearchCV,ParameterGrid

params = {
    'n_estimators': [100, 200, 300],
    
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10]
}

grid = GridSearchCV(RandomForestClassifier(), param_grid=params, cv=5)
grid.fit(X_train, y_train)
print("Best parameters:", grid.best_params_)

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
importances = pd.Series(best_model.feature_importances_, index=X.columns)
importances.sort_values().plot(kind='barh', figsize=(10,6), title="Feature Importances")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.2f}")
```
